```python
#Max Fibonacci Heap

class FibonacciHeap:
    class Node:
        def __init__(self,data):
            self.data = data
            self.parent = self.child = self.right = self.left = None
            self.mark = False
            self.degree = 0
            
    def __init__(self):
        self.head = None
        self.maxNode = None
        self.numberNodes = 0
        
    def insert(self, data):
        newNode = self.Node(data)
        newNode.left = newNode.right = newNode
        self.mergeWithHead(newNode)
        self.numberNodes += 1
            
    def mergeWithHead(self, newNode):
        if self.head is None:
            self.head = newNode
            self.maxNode = newNode
            #print "newnode"
            #print self.maxNode.data
        elif self.head.left == self.head:
            #print "2nd node"
            #print self.maxNode.data
            self.head.right.left = newNode
            newNode.right = self.head.right
            newNode.left = self.head
            self.head.right = newNode
            self.head.left = newNode
            if newNode.data > self.maxNode.data:
                self.maxNode = newNode
        else:
            #print "3rd node +"
            #print self.maxNode.data
            self.head.right.left = newNode
            newNode.right = self.head.right
            newNode.left = self.head
            self.head.right = newNode
            if newNode.data > self.maxNode.data:
                self.maxNode = newNode

    def removeChildFromParent(p,x):
        if p.child == p.child.right:
            p.child = None
        if p.child == x:
            p.child = x.right
            x.right.parent = p
        x.left.right = x.right
        x.right.left = x.left
        
    def cut(self, x, p):
        self.removeChildFromParent(p,x)
        p.degree -= 1
        self.mergeWithHead(x)
        x.parent = None
        x.mark = False
        
    def cascadingCut(self, y):
        p = y.parent
        if p is not None:
            if y.mark is False:
                y.mark = True
            else:
                self.cut(self, y, p)
                self.cascadingCut(p)
    
    def iterate(self, head):
        node = stop = head
        flag = False
        while True:
            if node == stop and flag is True:
                break
            elif node == stop:
                flag = True
            yield node
            node = node.right
            
    def removeFromHead(self, x):
        if self.head == x:
            self.head = x.right
        x.left.right = x.right
        x.right.left = x.left            
            
    def extractMax(self):
        m = self.maxNode
        if m is not None:
            if m.child is not None:
                children = [node for node in self.iterate(m.child)]
                for child in children:
                    self.mergeWithHead(child)
                    child.parent = None
                    child.mark = False
                    
            if m == m.right:
                self.head = None
                self.maxNode = None
            else:
                self.removeFromHead(m)
                self.maxNode = max([node for node in self.iterate(self.head)])
                self.pairwiseCombine()
            self.numberNodes -= 1
        return m
    
    def pairwiseCombine(self):
        nodes = [node for node in self.iterate(self.head)]
        #self.numberNodes = len(nodes)
        A = [None] * self.numberNodes
        for w in xrange(0, len(nodes)):
            x = nodes[w]
            d = x.degree
            print A[d]
            while A[d] != None:
                y = A[d] 
                if x.data < y.data:
                    #self.heapLink(y, x)
                    temp = x
                    x, y = y, temp
                self.heapLink(y, x)
                A[d] = None
                d += 1
            A[d] = x
        # find new min node - no need to reconstruct new root list below
        # because root list was iteratively changing as we were moving 
        # nodes around in the above loop
        for i in xrange(0, len(A)):
            if A[i] is not None:
                if A[i].data > self.maxNode.data:
                    self.maxNode = A[i]
        
    # actual linking of one node to another in the root list
    # while also updating the child linked list
    
    def pairwiseCombine2(self):
        nodes = [node for node in self.iterate(self.head)]
        #self.numberNodes = 1000
        A = [None] * self.numberNodes
        for w in xrange(0, len(nodes)):
            x = nodes[w]
            d = x.degree
            print A[d]
            while A[d] != None:
                y = A[d] 
                if x.data < y.data:
                    #temp = x
                    self.heapLink(x, y)
                    #x, y = y, x
                else:
                    self.heapLink(y, x)
                A[d] = None
                d += 1
            A[d] = x
        # find new min node - no need to reconstruct new root list below
        # because root list was iteratively changing as we were moving 
        # nodes around in the above loop
        for i in xrange(0, len(A)):
            if A[i] is not None:
                if A[i].data > self.maxNode.data:
                    self.maxNode = A[i]
        
    # actual linking of one node to another in the root list
    # while also updating the child linked list
    def mergeWithChild(self, parent, node):
        if parent.child is None:
            parent.child = node
            parent.degree += 1
        else:
            node.right = parent.child.right
            node.left = parent.child
            parent.child.right.left = node
            parent.child.right = node
            parent.degree += 1
            
    def heapLink(self, y, x):
        self.removeFromHead(y)
        y.left = y.right = y
        self.mergeWithChild(x, y)
        x.degree += 1
        y.parent = x
        y.mark = False
    
    def increaseKey(self,k,x):
        '''if k > x.data:
            if node.parent == None:
                self.removeFromHead(x)
            else:
                self.removeChildFromParent(x,x.parent)
        '''
        x.data += k
        y = x.parent
        if y is not None and x.data > y.data:
            self.cut(x, y)
            self.cascading_cut(y)
        if x.data > self.maxNode.data:
            self.maxNode = x
            
    
            
                    
                
        
            
            
        
        
```
